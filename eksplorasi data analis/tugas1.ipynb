<span style="color:blue">Import data dan Preview data</span>


```python
import pandas as pd #mengimport module pandas untuk manipulasi data
import openpyxl as op # menimport module openpyxl untuk membaca file exel

data =pd.read_excel(r"C:\Users\Michael Zidane\Downloads\matana project\Jumlah Kejadian Bencana Alam Menurut Provinsi 2021 new version.xlsx") # mengimport file excel ke jupyternotebook
df = pd.DataFrame(data) #membuat data frame

df.iloc[:, 0:5] #menampilkan preview data
```

<span style="color:pink">Split data menggunakan illoc</span>



```python
print("")
print("Split data Provinsi  Aceh")
print("")

data_split1 = df.iloc[0:1, 0:5]

print(data_split1)
print("")
print("Split data Provinsi Sumatera Utara")
print("")


data_split2 = df.iloc[1:2, 0:5]
print(data_split2)



data_split1.index = data_split2.index
```

<span style="color:red">Merge Data</span>


```python
df_merge = pd.merge(data_split1, data_split2, 
                    left_index=True, 
                    right_index=True , suffixes=('_Aceh','_Sumut'))

print("")
print("Hasil Merge")
print("")
print(df_merge)
```

<span style="color:yellow">Export To csv</span>


```python
df_merge.to_csv('output_csv.csv', sep=";", index=False)
```

```python
df_merge.to_excel('output_excel.xlsx', index=False)

```
