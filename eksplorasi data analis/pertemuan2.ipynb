### Membuat data frame

```python
# data frame dari dictionary
data_start = {
    "Name" : ["alice", "bob", "charlie"],
    "age" : [25, 30, 35],
    "city" :["new york", "los angels", "chigaco"]
}

df = pd.DataFrame(data_start)
print(df)
```

```python
#data frame dari list
data_list = [
    ["David", 28, "Miami"],
    ["Eva", 22, "Seatle"],
    ["Frank", 33, "Boston"]
]

df_list = pd.DataFrame(data_list, columns=["Name", "Age", "City"])
print(df_list)
```

```python
print (df_list.columns.to_list())
```

```python
df_list.columns = ["Nama", "Umur", "Kota"]
print(df_list)
```

```python
print(df_list["Nama"])
```

```python
print(df_list[["Nama", "Kota"]])
```

```python
# menampilkan baris 

print(df_list.iloc[1-1]) # untuk baris pertama
print(df_list.iloc[2-1]) # untuk baris kedua
```
