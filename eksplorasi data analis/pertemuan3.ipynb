### membuat data frame


```python
import pandas as pd

# Create a sample DataFrame

data = {'Product': ['A', 'B', 'A', 'B', 'C', 'C', 'A'],

'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Furniture', 'Furniture'],

'Sales': [250, 150, 350, 200, 100, 400, 300],

'Quantity': [10, 5, 7, 8, 3, 9, 4]}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.dtypes) #cek tipe data dalam dataframe
df.info() #cek data lengkap dalam dataframe
#ubah tipe data dalam dataframe
df["Sales"] = df["Sales"].astype(float)
print(df.dtypes)
```

```python
summary = df.groupby('Category').agg({
    "Sales" : "sum",
    "Quantity" : "mean",
})

print (summary)
```

```python
# Grouping data by Product and summarizing total sales

grouped = df.groupby('Product').agg({'Sales': 'sum', 'Quantity': 'mean'})

print(grouped)
```

```python
# Filtering rows where Sales are greater than 200

filtered = df[df['Sales'] > 200]

print(filtered)
```

```python
# Selecting Product and Sales columns

selected_columns = df[['Product', 'Sales']]

print(selected_columns)
```

```python
df['Revenue'] = None

print(df)
```

```python
# Sorting by Sales in descending order

sorted_df = df.sort_values('Sales', ascending=False)

print(sorted_df)
```
