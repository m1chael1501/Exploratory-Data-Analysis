### Student Exam Performance Dataset


###### Dataset ini berisikan 100 sampel murid yang mempunyai simulasi kinerja akademik siswa, termasuk informasi demografis, perilaku belajar, nilai mata pelajaran, dan hasil ujian akhir.

###### [Sumber data](https://www.kaggle.com/datasets/ssssws/student-exam-performance-dataset?resource=download)

#### Import dan Preview data

```python
import pandas as  pd
import openpyxl as op

data = pd.read_excel(r"C:\Users\Michael Zidane\Downloads\matana project\student stres 100.xlsx")
df = pd.DataFrame(data)

print(df)
```

#### Memfilter variabel yang hanya numerik saja

```python
df_numerik = df.select_dtypes(include=['number'])
print(df_numerik)
```

### Visualisai heatmap

```python
# membuat matriks korelasi

korelasi_matrix = df_numerik.corr()
print(korelasi_matrix)

# membuat visualisasi heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap (korelasi_matrix, annot = True, cmap='coolwarm')
plt.title('Heatmap Matriks Korelasi Student Exam performance')
plt.show()
```

### Visualisasi pairplot

```python
# menghitung matriks kovarians

kovarians_matrix = df_numerik.cov()
print (kovarians_matrix)

# membuat visualisasi pairplot

sns.pairplot(df_numerik)
plt.title("Pairplot Exam Performance Test")
plt.show()
```
