### Dataset Prediksi Harga Rumah


###### Dataset ini berisikan 50 data yang terkait tentang harga rumah, yang meliputi lokasi,tahun dibuat,kondisi,jumlah kamar,jumlah lantai dan garasi.Pada analisis ini ,terdapat dua metode regresi korelasi yaitu berganda dan sederhana.

###### [Sumber data](https://www.kaggle.com/datasets/zafarali27/house-price-prediction-dataset)

#### Import data dan membuat data frame

```python
import pandas as pd
import openpyxl as op

data = pd.read_excel(r"C:\Users\Michael Zidane\Downloads\matana project\Harga rumah.xlsx")
df = pd.DataFrame(data)

print(df)
```

#### Regresi Linear Berganda

```python
import pandas as pd
import statsmodels.api as sm

df_numerik = df.select_dtypes(include=['number'])
Y = df_numerik['Price']  
X = df_numerik.drop('Price', axis=1)
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

print(model.summary())
```

#### Cek asumsi Regresi Linear Berganda

Uji nomralitas residual

```python
from scipy.stats import shapiro

residual = model.resid

stat, p = shapiro(residual)
print("p-value:", p)
```

###### karena p-value lebih dari 0,05  berarti data tersebut aman

Uji Multikolinearitas (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["Variabel"] = X.columns
vif
```

###### karena VIF kurang dari 10 data tersebut aman

Uji Heteroskedastisitas

```python
from statsmodels.stats.diagnostic import het_breuschpagan

test = het_breuschpagan(residual, X)
p_value = test[1]

print("p-value:", p_value)
```

###### karena p value lebih dari 0,05 maka data tersebut homokedastis

Uji autokorelasi

###### Durbin-Watson = 2.397 (~2 →  tidak ada autokorelasi )


#### Regresi linear sederhana

```python
X1 = df_numerik[['Area']]
X1 = sm.add_constant(X1)

model_sederhana = sm.OLS(Y, X1).fit()
model_sederhana.summary()
```

#### Penjelasan metode regresi linear sederhana

###### Model regresi linear sederhana yang diperoleh adalah:

###### Price=585500−4.9853×Area

###### Model ini menunjukkan bahwa variabel Area memiliki hubungan negatif terhadap Price, di mana setiap kenaikan satu satuan luas area akan menurunkan harga sebesar 4.9853 satuan. Namun, berdasarkan hasil uji statistik, nilai p-value untuk variabel Area sebesar 0.885 (> 0.05), sehingga dapat disimpulkan bahwa variabel Area tidak berpengaruh signifikan terhadap Price.

###### Dengan demikian, model regresi linear sederhana ini belum mampu menjelaskan hubungan yang berarti antara Area dan Price.

#### Adjusted R-Square dan Interpretasinya

###### Nilai Adjusted R-Squared yang diperoleh adalah sebesar -0.020.

###### Nilai ini menunjukkan bahwa model regresi yang digunakan tidak mampu menjelaskan variasi dalam variabel Price. Bahkan, nilai Adjusted R-Squared yang negatif mengindikasikan bahwa model yang dibangun lebih buruk dibandingkan dengan model tanpa variabel independen (hanya menggunakan rata-rata).

###### Dengan demikian, dapat disimpulkan bahwa variabel Area tidak cukup kuat untuk menjelaskan perubahan pada Price, sehingga diperlukan variabel lain untuk membangun model yang lebih baik.
