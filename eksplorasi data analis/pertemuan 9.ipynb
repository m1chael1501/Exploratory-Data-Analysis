```python
import pandas as pd

data = pd.DataFrame({
    "A": [1, 2,None, 4, 5],
    "B": [5, None,None,8, 2],
    "C": [10, 20, 30, None, 50]
})

Row_deletion = data.dropna()
print("Data awal: ")
print(data)

print("Pembersihan baris data: ")
print(Row_deletion)
```

```python
print("Data awal; ")
print(data)

#menghapus kolom dengan missing data
Column_deletion = data.dropna(axis=1)
print("Pembersihan kolom data: ")
print(Column_deletion)

#axis 0 buat menghapus baris axis 1 buat menghapus kolom
```

```python
print(data)

#mengganti nilai yang hilang dengan mean
mean_imputed_data = data.fillna(data.mean())
print("Data setelah imputasi dengan mean: ")
print(mean_imputed_data)

#mengganti nilai yang hilang dengan median
median_imputed_data = data.fillna(data.median())
print("Data setelah imputasi dengan median: ")
print(median_imputed_data)

#mengganti nilai yang hilang dengan mode
mode_imputed_data = data.fillna(data.mode().iloc[0])
print("Data setelah imputasi dengan mode: ")
print(mode_imputed_data)
```

Forward fill

```python
print(data)

#menggunakan forward fill
ff_imputed_data = data.ffill()
print("Data setelah imputasi dengan forward fill: ")
print(ff_imputed_data)
```

backward fill

```python
print(data)
bf_imputed_data = data.bfill()
print("Data setelah imputasi dengan backward fill: ")
print(bf_imputed_data)
```

```python
print(data)

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)
knn_imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

print("Data setelah imputasi dengan KNN: ")

print(knn_imputed_data)
```

```python
print(data)

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
multivariate_imputed_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

print(multivariate_imputed_data)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

data_kategorik = pd.DataFrame({
    'Produk': ['Laptop', 'Smartphone', 'Tablet', None, 'Smartphone', 'Tablet', None],
    'Brand': ['Apple', 'Samsung', 'Apple', 'Samsung', None, 'Apple', 'Samsung']
})

# Menghapus baris yang memiliki missing values pada kolom 'Brand'
data_kategorik = data_kategorik.dropna(subset=['Brand'])

# Menghilangkan missing values untuk pelatihan decision tree
data_kategorik_clean = data_kategorik.dropna().copy()  # Gunakan .copy() untuk menghindari peringatan

# Label encoding untuk mengubah kategori menjadi angka
le_brand = LabelEncoder()
le_produk = LabelEncoder()

# Transformasi data kategorik menjadi numerik
data_kategorik_clean['Brand_encoded'] = le_brand.fit_transform(data_kategorik_clean['Brand'])
data_kategorik_clean['Produk_encoded'] = le_produk.fit_transform(data_kategorik_clean['Produk'])

# Membuat model Decision Tree Classifier
model = DecisionTreeClassifier()
X = data_kategorik_clean[['Brand_encoded']]  # Fitur (Brand yang sudah diencode)
y = data_kategorik_clean['Produk_encoded']   # Target yang akan diprediksi (Produk yang sudah diencode)
model.fit(X, y)

# Salinan data_kategorik untuk hasil prediksi
predicted_data_kategorik = data_kategorik.copy()

# Transformasi kolom 'Brand' yang hilang ke dalam bentuk numerik untuk prediksi
missing_data = predicted_data_kategorik[predicted_data_kategorik['Produk'].isna()].copy()  # Salin agar menghindari peringatan
missing_data['Brand_encoded'] = le_brand.transform(missing_data['Brand'])

# Prediksi missing values pada kolom 'Produk' (berdasarkan Brand)
predicted_produk = model.predict(missing_data[['Brand_encoded']])

# Decode hasil prediksi kembali ke kategori aslinya
predicted_data_kategorik.loc[predicted_data_kategorik['Produk'].isna(), 'Produk'] = le_produk.inverse_transform(predicted_produk)

# Menampilkan hasil prediksi dalam variabel baru tanpa overwrite data_kategorik asli
print("Data asli (data_kategorik):")
print(data_kategorik)

print("\nHasil prediksi dalam variabel baru (predicted_data_kategorik):")
print(predicted_data_kategorik)
```

```python
import pandas as pd

# Contoh data dengan format yang tidak konsisten
data_format = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Tanggal': ['2022-01-01', '01/02/2022', 'March 3, 2022', '2022-04-01'],
    'Pendapatan': ['10000', '$20000', '30000', '40,000']
})

# Menampilkan data asli
print("Data sebelum perbaikan format:")
print(data_format)

# Mengubah kolom 'Pendapatan' menjadi tipe numerik
data_format['Pendapatan'] = data_format['Pendapatan'].replace({r'\$': '', ',': ''}, regex=True).astype(float)

# Menstandarkan format tanggal
data_format['Tanggal'] = pd.to_datetime(data_format['Tanggal'])

# Menampilkan data setelah perbaikan
print("\nData setelah perbaikan format:")
print(data_format)
```

```python
import pandas as pd

# Contoh data dengan format yang tidak konsisten
data_format = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Tanggal': ['2022-01-01', '01/02/2022', 'March 3, 2022', '2022-04-01'],
    'Pendapatan': ['10000', '$20000', '30000', '40,000']
})

# Menampilkan data asli
print("Data sebelum perbaikan format:")
print(data_format)

# Mengubah kolom 'Pendapatan' menjadi tipe numerik
data_format['Pendapatan'] = data_format['Pendapatan'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Menstandarkan format tanggal
data_format['Tanggal'] = pd.to_datetime(data_format['Tanggal'], format='mixed', dayfirst=True)

# Menampilkan data setelah perbaikan
print("\nData setelah perbaikan format:")
print(data_format)
```

```python
#  Contoh data dengan nilai yang salah
data_invalid = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nama': ['Alice', 'Bob', 'Charlie', 'David'],
    'Umur': [25, -30, 35, 200]  # Nilai negatif dan umur yang tidak logis
})

# Menampilkan data asli
print("Data sebelum pembersihan nilai yang salah:")
print(data_invalid)

# Mengganti nilai yang salah dengan NaN
data_invalid['Umur'] = data_invalid['Umur'].apply(lambda x: x if 0 < x < 120 else None)

print("\nData setelah pembersihan nilai yang salah:")
print(data_invalid)
```
