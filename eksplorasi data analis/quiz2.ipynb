```python
import pandas as pd #mengimport module pandas untuk manipulasi data
import openpyxl as op # menimport module openpyxl untuk membaca file exel

data =pd.read_excel(r"C:\Users\Michael Zidane\Downloads\matana project\Jumlah Kejadian Bencana Alam Menurut Provinsi 2021 new version.xlsx") # mengimport file excel ke jupyternotebook
df = pd.DataFrame(data) #membuat data frame

df.iloc[:, 0:5] #menampilkan preview data
```

```python
import matplotlib.pyplot as plt

df_10 = df.head(10)

plt.figure(figsize=(10,6))

plt.bar(df_10["Provinsi"], df_10["Banjir"])

plt.title("Jumlah Kejadian Banjir")
plt.xlabel("Provinsi")
plt.ylabel("Jumlah Banjir")

plt.xticks(rotation=45)  # supaya nama provinsi miring biar kebaca
plt.show()
```
