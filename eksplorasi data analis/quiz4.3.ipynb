### Analisis korelasi dan Kovariansi


#### Analisis kovarians matriks
###### Ukuran statistik yang menggambarkan derajat arah hubungan linear antar dua variabel

```python
import pandas as pd

#contoh data untuk praktek

data = {'A' : [1, 2 , 3 ,4 ,5],
'B' : [5, 4 , 3 , 2, 1],
'C': [2, 3 ,4 ,5 ,6]}

# menghitung matriks kovarians

df = pd.DataFrame(data)
covariance_matrix = df.cov()
print (covariance_matrix)

# menghitung matriks korelasi
correlation_matrix = df.corr()
print(correlation_matrix)
```

#### Visualisasi  dari corelations

###### Dua visualisasi umum yang sering digunakan adalaha heatmap dan pairplot.Heatmap adalah representasi matriks korelasi menggunakan warna untuk menunjukkan kekuatan hubungan antar variabel.Pairplot adalah kumpulan scatter plot yang memvisualisasikan hubungan antar setiap pasangan variabel

```python
# membuat heatmap di python
import seaborn as sns
import matplotlib.pyplot as plt

# membuat heatmap dari matriks korelasi
sns.heatmap (correlation_matrix, annot = True, cmap='coolwarm')
plt.title('Heatmap Matriks Korelasi')
plt.show()
```

```python
# membuat pairplot di python

import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)
plt.title('Pair Plot Data')
plt.show()
```
